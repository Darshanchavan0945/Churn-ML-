```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Data Cleaning

```python
ch.drop('customerID',axis=1,inplace=True)
```

```python
ch['TotalCharges'] = ch['TotalCharges'].str.replace(',', '') 
ch['TotalCharges'] = pd.to_numeric(ch['TotalCharges'], errors='coerce') 
```

```python
ch.isnull().sum()[ch.isnull().sum()>0]
```

```python
ch.TotalCharges.fillna(ch.TotalCharges.mean(),inplace=True)
```

```python
ch.select_dtypes(include='object').columns
```

```python
ch.StreamingMovies.unique()
```

```python
ch.gender.replace({'Female':0, 'Male':1},inplace=True)
ch.Partner.replace({'Yes':1, 'No':0},inplace=True)
ch.Dependents.replace({'Yes':1, 'No':0},inplace=True)
ch.PhoneService.replace({'Yes':1, 'No':0},inplace=True)
ch.MultipleLines.replace({'No phone service':0, 'No':0, 'Yes':1},inplace=True)
ch.InternetService.replace({'DSL':1, 'Fiber optic':2, 'No':0},inplace=True)
ch.OnlineSecurity.replace({'No internet service':0, 'No':0, 'Yes':1},inplace=True)
ch.OnlineBackup.replace({'No internet service':0, 'No':0, 'Yes':1},inplace=True)
ch.DeviceProtection.replace({'No internet service':0, 'No':0, 'Yes':1},inplace=True)
ch.TechSupport.replace({'No internet service':0, 'No':0, 'Yes':1},inplace=True)
ch.StreamingTV.replace({'No internet service':0, 'No':0, 'Yes':1},inplace=True)
ch.StreamingMovies.replace({'No internet service':0, 'No':0, 'Yes':1},inplace=True)
ch.Contract.replace({'Month-to-month':1, 'One year':2, 'Two year':3},inplace=True)
ch.PaperlessBilling.replace({'Yes':1, 'No':0},inplace=True)
ch.PaymentMethod.replace({'Electronic check':1, 'Mailed check':2, 'Bank transfer (automatic)':3,
       'Credit card (automatic)':4},inplace=True)
ch.Churn.replace({'Yes':1, 'No':0},inplace=True)
```

```python
ch.head(2)
```

```python
ch.info()
```

# Sampling

```python
from sklearn.model_selection import train_test_split
ch_train,ch_test=train_test_split(ch,test_size=.2)
```

```python
ch_train.shape
```

```python
ch_test.shape
```

```python
# create x and y
```

```python
ch_train_x = ch_train.drop(columns=['Churn'],axis=1)
ch_train_y = ch_train['Churn']
ch_test_x = ch_test.drop(columns=['Churn'],axis=1)
ch_test_y = ch_test['Churn']
```

```python
ch_test_x.head()
```

# Logistic Model

```python
from sklearn.linear_model import LogisticRegression
logreg =LogisticRegression()
```

```python
logreg.fit(ch_train_x,ch_train_y)
```

```python
logref_pred_train=logreg.predict(ch_train_x)
logref_pred_test=logreg.predict(ch_test_x)
```

```python
from sklearn.metrics import confusion_matrix ,accuracy_score ,precision_score,recall_score
```

```python
confusion_matrix(ch_test_y,logref_pred_test)
```

```python
accuracy_score(ch_test_y,logref_pred_test)
```

```python
precision_score(ch_test_y,logref_pred_test)
```

```python
recall_score(ch_test_y,logref_pred_test)
```
